## Ciencia de datos
</div>

Con Colab, puedes aprovechar toda la potencia de las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código de abajo utiliza <strong>NumPy</strong> para generar datos aleatorios y <strong>Matplotlib</strong> para visualizarlos. Para editar el código, solo tienes que hacer clic en la celda.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
dfcausa = pd.read_excel("/content/Causa.xlsx")
dfriesgo= pd.read_excel("/content/RIESGO.xlsx")
```

```
dfcausa = dfcausa.applymap(lambda s: s.lower() if type(s) == str else s)
dfcausa
```

```
dfriesgo = dfriesgo.applymap(lambda s: s.lower() if type(s) == str else s)
dfriesgo
```

```
def preprocess_text(text):
    stop_words = set(stopwords.words('spanish'))  # Now stopwords is defined
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(str(text).lower())  # Tokenización y conversión a minúsculas
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha()]  # Lematización y eliminación de números/símbolos
    tokens = [word for word in tokens if word not in stop_words]  # Remover stopwords
    return ' '.join(tokens)

dfcausa['causa_procesado'] = dfcausa["Causa"].apply(preprocess_text)
```

```
dfcausa
```

```
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.80,
    min_df=10,
    max_features=500,
)
```

```
causasTF = dfcausa["causa_procesado"]
tfidf_matrix = tfidf_vectorizer.fit_transform(causasTF)

```

```
from sklearn.metrics import silhouette_score
silhouette_scores = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(tfidf_matrix)
    silhouette_scores.append(silhouette_score(tfidf_matrix, labels))

# Graficar la puntuación Silhouette
plt.figure(figsize=(8, 4))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Puntuación Silhouette para diferentes valores de k')
plt.xlabel('Número de Clústeres (k)')
plt.ylabel('Puntuación Silhouette')
plt.show()
```

```
# Paso 1: Preprocesamiento de datos (limpiar y eliminar valores nulos)
data_clean = dfcausa.dropna(subset=["causa_procesado"])  # Eliminar valores nulos en la columna "Causa"
textos_causa = dfcausa["causa_procesado"].str.strip()  # Eliminar espacios en blanco al inicio y al final

# Paso 2: Configurar y transformar los textos con TF-IDF
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.80,
    min_df=10,
    max_features=500,
)
tfidf_matrix = tfidf_vectorizer.fit_transform(textos_causa)

# Paso 3: Determinar el número óptimo de clústeres con Silhouette
silhouette_scores = []
k_range = range(2, 11)  # Probar entre 2 y 10 clústeres

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(tfidf_matrix)
    silhouette_scores.append(silhouette_score(tfidf_matrix, labels))

# Paso 4: Aplicar clustering con el número óptimo de clústeres (determinado por gráfica o heurística)
num_clusters = 3  # Valor inicial, puede ajustarse tras analizar los resultados
kmeans_final = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
data_clean["Cluster"] = kmeans_final.fit_predict(tfidf_matrix)
```

```
data_clean
```

```
silhouette_avg = silhouette_score(tfidf_matrix, kmeans_final.labels_)
print(f"Puntuación Silhouette para {num_clusters} clústeres: {silhouette_avg}")
```

```
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
tfidf_pca = pca.fit_transform(tfidf_matrix.toarray())
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    tfidf_pca[:, 0], tfidf_pca[:, 1], c=data_clean["Cluster"], cmap='viridis', s=50, alpha=0.7
)
plt.colorbar(scatter, label='Clúster')
plt.title("Visualización de Clústeres (PCA)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(alpha=0.3)
plt.show()
```
